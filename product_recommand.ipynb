```python
import numpy as np
dataset_filename = "affinity_dataset.txt"
X = np.loadtxt(dataset_filename)
n_samples, n_features = X.shape
print("This dataset has {0} samples and {1} features".format(n_samples, n_features))
```

```python
print(X)
```

```python
# The names of the features, for your reference.
features = ["bread", "milk", "cheese", "apples", "bananas"]
```

```python
#X->Y

from collections import defaultdict

valid_rules = defaultdict(int)
invalid_rules = defaultdict(int)
num_occurences = defaultdict(int)

for sample in X:
    for premise in range(n_features):
        if sample[premise] == 0: continue
        # Record that the premise was bought in another transaction
        num_occurences[premise] += 1
        for conclusion in range(n_features):
            if premise == conclusion:  
                continue
            if sample[conclusion] == 1:
                # This person also bought the conclusion item
                valid_rules[(premise, conclusion)] += 1
            else:
                # This person bought the premise, but not the conclusion
                invalid_rules[(premise, conclusion)] += 1
support = valid_rules
confidence = defaultdict(float)
for premise, conclusion in valid_rules.keys():
    confidence[(premise, conclusion)] = valid_rules[(premise, conclusion)] / num_occurences[premise]
```

```python
for premise, conclusion in confidence:
    premise_name = features[premise]
    conclusion_name = features[conclusion]
    print("Rule: If a person buys {0} they will also buy {1}".format(premise_name, conclusion_name))
    print(" - Confidence: {0:.3f}".format(confidence[(premise, conclusion)]))
    print(" - Support: {0}".format(support[(premise, conclusion)]))
    print("")
```

```python
#X,Y->Z

from collections import defaultdict

valid_rules = defaultdict(int)
invalid_rules = defaultdict(int)
num_occurences = defaultdict(int)

for sample in X:
    for premise1 in range(n_features):
        if sample[premise1] == 0: continue
    
        for premise2 in range(n_features):
            if premise1 == premise2:  
                continue
            if sample[premise2] == 1:
                num_occurences[premise1,premise2]+=1
                for conclusion in range(n_features):
                    if premise1 == conclusion or premise2==conclusion: 
                        continue
                    if sample[conclusion] == 1:
                        # This person also bought the conclusion item
                        valid_rules[(premise1,premise2, conclusion)] += 1

                    else:
                        # This person bought the premise, but not the conclusion
                        invalid_rules[(premise1,premise2, conclusion)] += 1
support = valid_rules
confidence = defaultdict(float)
for premise1,premise2, conclusion in valid_rules.keys():
    confidence[(premise1,premise2, conclusion)] = valid_rules[(premise1,premise2, conclusion)] / num_occurences[premise1,premise2]
```

```python
for premise1, premise2, conclusion in confidence:
    premise_name1 = features[premise1]
    premise_name2 = features[premise2]
    conclusion_name = features[conclusion]
    print("Rule: If a person buys {0},{1} they will also buy {2}".format(premise_name1,premise_name2, conclusion_name))
    print(" - Confidence: {0:.3f}".format(confidence[(premise1,premise2, conclusion)]))
    print(" - Support: {0}".format(support[(premise1,premise2, conclusion)]))
    print("")
```

```python
#W,X,Y->Z

from collections import defaultdict

valid_rules = defaultdict(int)
invalid_rules = defaultdict(int)
num_occurences = defaultdict(int)

for sample in X:
    for premise1 in range(n_features):
        if sample[premise1] == 0: continue
        for premise2 in range(n_features):
            if premise1 == premise2:  
                continue
            if sample[premise2] == 1:
                for premise3 in range(n_features):
                    if premise3 == premise2 or premise3 == premise1: continue
                    if sample[premise3] == 1:
                        num_occurences[premise1,premise2,premise3]+=1
                        for conclusion in range(n_features):
                            if premise1 == conclusion or premise2==conclusion or premise3 == conclusion:  
                                continue
                            if sample[conclusion] == 1:
                                # This person also bought the conclusion item
                                valid_rules[(premise1,premise2,premise3, conclusion)] += 1

                            else:
                                # This person bought the premise, but not the conclusion
                                invalid_rules[(premise1,premise2,premise3, conclusion)] += 1
support = valid_rules
confidence = defaultdict(float)
for premise1,premise2,premise3, conclusion in valid_rules.keys():
    confidence[(premise1,premise2,premise3, conclusion)] = valid_rules[(premise1,premise2,premise3, conclusion)] / num_occurences[premise1,premise2,premise3]
```

```python
for premise1, premise2,premise3, conclusion in confidence:
    premise_name1 = features[premise1]
    premise_name2 = features[premise2]
    premise_name3 = features[premise3]
    conclusion_name = features[conclusion]
    print("Rule: If a person buys {0}, {1}, {2} they will also buy {3}".format(premise_name1,premise_name2,premise_name3, conclusion_name))
    print(" - Confidence: {0:.3f}".format(confidence[(premise1,premise2,premise3, conclusion)]))
    print(" - Support: {0}".format(support[(premise1,premise2,premise3, conclusion)]))
    print("")
```

```python
#v,w,x,y=>z

from collections import defaultdict

valid_rules = defaultdict(int)
invalid_rules = defaultdict(int)
num_occurences = defaultdict(int)

for sample in X:
    for premise1 in range(n_features):
        if sample[premise1] == 0: continue
        for premise2 in range(n_features):
            if premise1 == premise2:  
                continue
            if sample[premise2] == 1:
                for premise3 in range(n_features):
                    if premise3 == premise2 or premise3 == premise1: continue
                    if sample[premise3] == 1:
                        for premise4 in range(n_features):
                            if premise4 ==  premise3 or premise4 == premise2 or premise4==premise1: continue
                            if sample[premise4]==1:
                                num_occurences[premise1,premise2,premise3,premise4]+=1
                                for conclusion in range(n_features):
                                    if premise1 == conclusion or premise2==conclusion or premise3 == conclusion or premise4==conclusion:  
                                        continue
                                    if sample[conclusion] == 1:
                                            # This person also bought the conclusion item
                                        valid_rules[(premise1,premise2,premise3,premise4, conclusion)] += 1
                                    else:
                                            # This person bought the premise, but not the conclusion
                                        invalid_rules[(premise1,premise2,premise3,premise4, conclusion)] += 1
support = valid_rules
confidence = defaultdict(float)
for premise1,premise2,premise3,premise4, conclusion in valid_rules.keys():
    confidence[(premise1,premise2,premise3,premise4, conclusion)] = valid_rules[(premise1,premise2,premise3,premise4, conclusion)] / num_occurences[premise1,premise2,premise3,premise4]
```

```python
for premise1, premise2,premise3,premise4, conclusion in confidence:
    premise_name1 = features[premise1]
    premise_name2 = features[premise2]
    premise_name3 = features[premise3]
    premise_name4 = features[premise4]
    conclusion_name = features[conclusion]
    print("Rule: If a person buys {0}, {1}, {2}, {3} they will also buy {4}".format(premise_name1,premise_name2,premise_name3,premise_name4, conclusion_name))
    print(" - Confidence: {0:.3f}".format(confidence[(premise1,premise2,premise3,premise4, conclusion)]))
    print(" - Support: {0}".format(support[(premise1,premise2,premise3,premise4, conclusion)]))
    print("")
```
